## Два интерфейса

#### В стиле MATLAB

```python
plt.figure() # создаем рисунок графика

# Создаеи реовую из двух областей графика и задаем текущую ось
plt.subplot(2, 1, 1)  # (rows, columns, panel number)
plt.plot(x, np.sin(x))
# Создаем вторую область и задаем текущую ось
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

# Интерфейс сохраняет состояние, отслеживая текуший рисунок и его оси координат, и выполняет для него все команды plt
# Получить на них ссылки можно с помощью команд:
# plt.gcf() - get current figure
# plt.gca() - get current axes(оси координат)

# Хотя в случае полстых графиков этот интерфейс с сохранением состояния быстр и удобен,
# его использование может привести к проблемам. Например после создания второй области, 
# довольно трудно и громоздко вернуться к форматированию первой.
```

#### Объектно-ориентированный интерфейс

```python
# Подходит для более сложных ситуаций, когда нужно больше возоможностей управления рисунком.
# В этом интерфейсе не полагаются на понятие текущего рисунка или осей,
# а являются методами, явным образом определяемыз объектов Figure и Axes.
# Перерисуем предыдущий рисунок с его помощью.

# Сначала создаем сетку графикаов
# ax будет массивом из двух объктов
fig, ax = plt.subplots(2)

# Вызываеи метод plot() соответствующего объекта
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

## Настройка графика: цвета и линий

```python
# Для настройки цвета используется слово color
# Цвет можно передавать разными способами
plt.plot(x, np.sin(x - 0), color='blue')        # Задаем цвет по названию
plt.plot(x, np.sin(x - 1), color='g')           # Краткий код цвета (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Шкала оттенков серого цвета, значение в диапазоне от 0 до 1
plt.plot(x, np.sin(x - 3), color='#DDFF44')     # 16-ричный код (RRGGBB) om 00 до FF
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # Кортеж RGB, значения от 0 до 1
plt.plot(x, np.sin(x - 5), color='chartreuse');  # поддерживаются все названия HTML цветов
# Если цвет не зодане библиотека Matplotlib автоматически будет перебирать цвета
```

```python
# Стиль линий можно настраивать с помощью ключевого слова linestyle
plt.plot(x, x + 0, linestyle='solid')    
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
```

```python
# Можно использовать сокращенные коды
plt.plot(x, x + 0, linestyle='-')     # Сплошная линия
plt.plot(x, x + 1, linestyle='--')    # Штриховая линия
plt.plot(x, x + 2, linestyle='-.')    # Штрихпунктирная линия
plt.plot(x, x + 3, linestyle=':');    # Пунктирная линия
```

```python
# Можно объединить задание цвета и стиля линий в одном ключевом аргументе
plt.plot(x, x + 0, '-g')     # Зеленая сплошная линия
plt.plot(x, x + 1, '--c')    # Голубая штриховая линия
plt.plot(x, x + 2, '-.k')    # Черная штрихпунктирная линия
plt.plot(x, x + 3, ':r');    # Красная пунктирная линия
```

## Простые диаграммы рассеяния

### Построение диаграмм рассения с помощью plt.plot() / ax.plot()

```python
# Точки не соединяются отрезками
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');  # Третий аргумент описывает описывает тип символа
```

```python
# Большинство точечных символов интуитивно понятны, продемонстрируем часто используемые
rng = np.random.RandomState(0)
for marker in ['o', '.', ',','x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label='marker="{}"'.format(marker))
plt.legend(numpoints=1)   # хз что дал numpoints
plt.xlim(0, 1.8);
```

```python
# Эти симовольные коды можно использовать совместно с кодами линий и цветов
plt.plot(x, y, '-ok');     # линия(-), маркер круга(о), черный цвет(k)
```

```python
# С помощью дополнительных ключевых параметров можно задавать множество свойств линий и маркеров
plt.plot(x, y, '-o', color='gray', markersize=10, linewidth=3, markerfacecolor='white', markeredgecolor='green', markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

### Построение диаграмм рассеяния с помощью функции plt.scatter()

```python
# Еще большими возможностями обладаеи метод для построения диаграмм рассеяния plt.scatter()
plt.scatter(x, y, marker=',');
```

```python
# Основное отличие состоит в том, что можно с помощью plt.scatter() можно создавать диаграммы рассеяния с индивидуально
# задаваемыми(или выбираемыми в соответствии с данными) свойствами каждой точки(размер, цвет заливки, цвет рамки)

# Создадим случайную диаграмму рассеяния
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')  # alpha - задает уровень прозрачности
plt.colorbar();    # Отображение цветовой шкалы

# Можно использовать цвет и размер для передачи информации с целью иллюстрации многомерных данных
```

### Непрерывные погрешности

```python
# скомбинируем plt.plot и plt.fill_between
# Выполним для примера простую регрессию на основе Гауссова процесса.

from sklearn.gaussian_process import GaussianProcess

# Описываем модель и отрисовываем некоторые данные
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# Выполняем подгонку Гауссова процусса
gp = GaussianProcess(corr='cubic', theta0=1e-2, thetaL=1e-4, thetaU=1E-1, random_start=100)
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], eval_MSE=True)
dyfit = 2 * np.sqrt(MSE) # 2*сигма область с уровнем доверия 95%
```

```python
# Визуализируем результат
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit, color='gray', alpha=0.2)
plt.xlim(0, 10);
```
