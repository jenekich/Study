# Наивная байесовская классификация

Различные виды наивных байесовских классификаторов основываются на различных "наивных" допущениях относительно данных, рассмотрим несколько из них.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(style='darkgrid', font_scale=1.2)
```

### Гауссов наивный байесовкий классификатор

Один из самых простых классификаторов - Гауссов. В этом классификаторе допущение состоит в том, что *данные всех категорий взяты из простого нормального распределения*.

```python
sns.set_style('darkgrid', {'axes.facecolor': '.85'})
```

```python
# Создадим данные
from sklearn.datasets import make_blobs
X, y = make_blobs(100, 2, centers=3, random_state=2, cluster_std=1)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis');
y
```

Один из способов создания простой модели - допущение о том, что данные подчиняются нормальному распределению без ковариации между измерениями. Для обучения такой модели достаточно найти среднее значение и стандартное отклонение точек внутри каждой из категорий.

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, y);
```

```python
# Создадим данные метки для которых будем предсказывать
rng = np.random.RandomState(0)
Xnew = [-6, -14] + [14, 18] * rng.rand(2000, 2)
ynew = model.predict(Xnew)
```

```python
# Предсказываем метки точек
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis')
lim = plt.axis()
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=20, cmap='viridis', alpha=0.2)
plt.axis(lim);
```

Положительная сторона этого байесовского формального метода заключается в возможности естественной вероятностной классификации, рассчитывать которую можно с помощью метода predict_proba:

```python
yprob = model.predict_proba(Xnew)
yprob[-8:].round(2)
```

Столбцы отражают вероятность каждой метки по данному классификатору.

#### Наивный байесовский классификатор имеет смысл использовать в следующих случаях

В силу строгих допущений отностиельно данных наивные байесовские классификаторы обычно работают хуже, чем более сложные модели. Тем не менее у них есть нексолько достоинств:
* они выполняют как обучение, так и предсказание исключительно быстро;
* обеспечивает простое вероятностное предсказание;
* их результаты часто очень легки для интерпретации;
* у них очень мало (если вообще есть) настраиваемых параметров.

Эти достоинства означают, что наивный байсовский классификатор зачастую оказывается удачным выбором для первоначальной классификации. Если он демонстрирует удовлетворительные результаты, то можно на нем и остановиться. Если же нет, то всегда можно начать использовать более сложные модели, имея первоначальную представление.

Наивные байсовские классификаторы склонны демонстрировать особенно хорошие результаты в следующих случаях:
* когда данные действительно соответствуют наивным допущениям(бывает очень редко)
* для очень хорошо разделяемых категорий, когда сложность модели не так важна
* для данных с очень большим числом измерений, когда сложность модели не столь важна

Два последних случая взаимосвязаны, т.к. с ростом кол-ва измерений у набора данных вероятность близости любых двух точек падает. Это значит, что кластеры в многомерных случаях имеют склонность к изоляции. Поэтому упрощенные классификаторы при росте числа изменрений начинают работать не хуже, а то и лучше более сложных моделей, когда данных достаточно много даже простая модель может оказаться эффективной.
