**Случайные леса** - это ансамбль из более простых моделей деревьев принятия решений. Интересно, что результат голосования среди достаточно большого количества оценивателей может оказаться лучше результата любого из отдельных оценивателей.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

Начнем просто с деревьев принятия решения.

```python
# Первые несколько шагов дерева принятия решений
# Если в ячейке только один цвет, то больше она не делится.
from helpers_05_08 import visualize_tree           # дополнительный py-файл с вспомогательными функциями для вищзуализации
from sklearn.tree import DecisionTreeClassifier    # Процесс обучения дерева принятия решений
from sklearn.datasets import make_blobs


fig, ax = plt.subplots(1, 4, figsize=(16, 3))
fig.subplots_adjust(left=0.02, right=0.98, wspace=0.1)

X, y = make_blobs(n_samples=300, centers=4,
                  random_state=0, cluster_std=1.0)

for axi, depth in zip(ax, range(1, 5)):
    model = DecisionTreeClassifier(max_depth=depth)
    visualize_tree(model, X, y, ax=axi)
    axi.set_title('Глубина = {0}'.format(depth))
```

```python
# Создадим функцию, которая будет рисовать результат работы нашей модели
def visualize_classifier(model, X, y, ax=None, cmap='rainbow'):
    ax = ax or plt.gca()
    
    # Рисуем точки
    ax.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=cmap,
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # Обучаемся
    model.fit(X, y)
    xx, yy = np.meshgrid(np.linspace(*xlim, num=200),
                         np.linspace(*ylim, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    # Создаем цветной график с результатами
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap=cmap, zorder=1)

    ax.set(xlim=xlim, ylim=ylim)
```

### Ансамбль

Идея комбинации нескольких переобученных оценивателей для снижения эффекта этого переобучения лежит в основе метода ансамбля под названием **bagging**. Баггинг использует ансамбль паралельно работающих переобученных оценивателей и усредняет результаты для получения оптимальной классификации. Ансамбль случайных деревьев принятия решений называют **случайным лесом(random forest)**.

```python
# Выполнить такую баггинг классификацию можно вручную:
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier

tree = DecisionTreeClassifier()
bag = BaggingClassifier(tree, n_estimators=100, max_samples=0.8,
                        random_state=1)

bag.fit(X, y)
visualize_classifier(bag, X, y);
```

В этом примере мы рандомизировали данные путем обучения оценивателей на случайном подмножестве, состоящем из 80% точек. На практике для более эффективной рандомизации деревьев обеспечивается определенная стохастичность процесса выбора разбиений и обучение каждый раз происходит на всех данных.

```python
# подобный оптимизированный ансамбль случайных деревьев, автоматически выполняющий всю рандомизацию, реализован тут:
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=0)
visualize_classifier(model, X, y);
```

### Регрессия с помощью случайных деревьев

Случайные леса могут быть также полезны для регрессии.

```python
# Например у нас есть данные, полученные из сочетания быcтрых и медленных колебаний, а также шума.
rng = np.random.RandomState(42)
x = 10 * rng.rand(200)

def model(x, sigma=0.3):
    fast_oscillation = np.sin(5 * x)
    slow_oscillation = np.sin(0.5 * x)
    noise = sigma * rng.randn(len(x))

    return slow_oscillation + fast_oscillation + noise

y = model(x)
plt.errorbar(x, y, 0.3, fmt='o');
```

```python
# Найдем оптимальную апроксимирующую кривую с помощью регрессии на основе случайного леса
# В этом случае используется RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
fig = plt.figure(figsize=(8, 6))
forest = RandomForestRegressor(200)
forest.fit(x[:, None], y)

xfit = np.linspace(0, 10, 1000)
yfit = forest.predict(xfit[:, None])
ytrue = model(xfit, sigma=0)

plt.errorbar(x, y, 0.3, fmt='o', alpha=0.5)
plt.plot(xfit, yfit, '-r');
plt.plot(xfit, ytrue, '-k', alpha=0.5);
```

Как мы видим регрессор на основание случайного леса достаточно гибка для апроксимации мультипериодических данных, без необходимости использования мультипериодической модели.
